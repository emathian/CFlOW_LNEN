### LCNEC Data set

```python
rootdir =  '/gpfsscratch/rech/ohv/ueu39kt/KI67_Tumoral_Normal_V2'
tneid_selected = ['TNE2123', 'TNE2128', 'TNE2130', 'TNE2125', 'TNE1536', 'TNE1560']
test_tiles_list = []
train_tiles_list = []
for folder in os.listdir(rootdir):
    for tne_select in tneid_selected:
        if folder.find(tne_select) != -1:
            for tile in os.listdir(os.path.join(rootdir, folder, "Normal")):
                test_tiles_list.append(os.path.join(rootdir, folder, "Normal", tile))
            # List tumoral 
            tumor_tiles =  os.listdir(os.path.join(rootdir, folder, "Tumoral"))
            random.shuffle(tumor_tiles)
            for tile in tumor_tiles[:150]:
                train_tiles_list.append(os.path.join(rootdir, folder, "Tumoral", tile))
            for tile in tumor_tiles[150:]:
                test_tiles_list.append(os.path.join(rootdir, folder, "Tumoral", tile))
```

```python
with open('/gpfsscratch/rech/ohv/ueu39kt/KI67_Tumoral_Normal_V2/LCNECLike_train_tiles_list.txt', 'w') as f:
    for ele in train_tiles_list:
        f.write(ele+'\n')
```

```python
with open('/gpfsscratch/rech/ohv/ueu39kt/KI67_Tumoral_Normal_V2/LCNECLike_test_tiles_list.txt', 'w') as f:
    for ele in test_tiles_list:
        f.write(ele+'\n')
```

```python
rootdir =  '/gpfsscratch/rech/ohv/ueu39kt/KI67_individual_data_for_segmentation'
for TNEID in os.listdir(rootdir):
    test_tiles_list = []
    train_tiles_list = []
    for tile in os.listdir(os.path.join(rootdir, TNEID, "Normal")):
                test_tiles_list.append(os.path.join(rootdir, TNEID, "Normal", tile))
    tumor_tiles =  os.listdir(os.path.join(rootdir, TNEID, "Tumoral"))
    random.shuffle(tumor_tiles)
    for tile in tumor_tiles[:150]:
        train_tiles_list.append(os.path.join(rootdir, TNEID, "Tumoral", tile))
    for tile in tumor_tiles[150:]:
        test_tiles_list.append(os.path.join(rootdir, TNEID, "Tumoral", tile))
    
    with open(f'{rootdir}/{TNEID}_train_tiles_list.txt', 'w') as fr:
        for ele in train_tiles_list:
            fr.write(ele+'\n')
    fr.close()
    with open(f'{rootdir}/{TNEID}_test_tiles_list.txt', 'w') as f:
        for ele in test_tiles_list:
            f.write(ele+'\n')
    f.close()


```
